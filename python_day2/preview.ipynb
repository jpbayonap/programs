```python
import matplotlib.pyplot as plt
import numpy as np
#
# Control size  by default it takes the least element size (set it always)
fig = plt.figure(figsize =(8,8))
plt.title('Pie chart',size =15)
#  labels
labels =['Python','Java','C++','Ruby']

# 
values = [20,30,40,10]

#
colors =['orange', 'red', 'blue', 'magenta']

# explode any  the labels
explode = [0,0.3,0,0]

#  shadow  
#show the ratios as percentage  autopct %
auto = '%0.1f%%'

# distance : autopct
dist = 0.4

# radius
r =1.5
# set labels text size textprops
tsize = {'fontsize':20}
plt.pie(values, labels=labels, colors=colors, shadow=True,
        explode=explode, autopct=auto, pctdistance=dist,
        radius=r, textprops=tsize)

plt.show()

```

```python
#
import matplotlib.pyplot as plt
import numpy as np
#
# Control size  by default it takes the least element size (set it always)
fig = plt.figure(figsize =(6,6))
plt.title('Pie chart',size =15)
#  labels
labels =['Python','Java','C++','Ruby']

# 
values = [20,30,40,10]

#
colors =['orange', 'red', 'blue', 'magenta']

# wedgeprops ={'linewidth':, 'width': ,'edgecolor': }
#linewidth 
#width
#edgecolor
exp = [0.1,0,0,0]
auto = '%0.1f%%'
rad = 2
wp = {'linewidth':4, 'width':0.6,'edgecolor':'green'}
#textprops 
ftsize = {'fontsize':16,'color':'purple', 'weight':'bold'}
pdist = 0.8
rot = 90 
plt.pie(values,labels=labels, colors=colors, 
        shadow=True, explode=exp, autopct=auto,
        pctdistance=pdist, radius=rad, wedgeprops=wp,
        startangle=rot, textprops=ftsize)
plt.show()
```

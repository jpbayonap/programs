```python
# 対数スケール
import pandas as pd
import matplotlib.pyplot as plt
world = pd.read_csv('world2015.csv')
size = world.Population/ 1e6  # one million
plt.xscale('log') # X軸に対数を取る
plt.xlim(world.GDP_per_capita.min(),world.GDP_per_capita.max())
plt.scatter(world.GDP_per_capita, world.Life_expectancy, s=size, alpha=0.5)
plt.show()
```

```python
# 色を付ける
import pandas as pd
import matplotlib.pyplot as plt
world = pd.read_csv('world2015.csv')
'''map 関数は受け取った数字を特定の関数に渡して実行する'''
map_dict = {
    'Asia':'green',
    'Europe':'red',
    'Africa': 'yellow',
    'North America':'blue',
    'South America':'blue',
    'Oceania':'c'
}
#　worls csv　中の Continent コラムの要素をそれぞれの色に変換する
colors = world.Continent.map(map_dict)
size = world.Population/ 1e6  # one million
plt.xscale('log') # X軸に対数を取る

plt.xlim(world.GDP_per_capita.min(),world.GDP_per_capita.max())
plt.scatter(world.GDP_per_capita, world.Life_expectancy,c=colors, s=size, alpha=0.5)
plt.show()
```

```python
#人口と人当たりgdpの関係の散布図
import matplotlib.pyplot as plt
import pandas as pd
world = pd.read_csv('world2015.csv')

fig = plt.figure(figsize=(15,8))
map_dict = {
    'Asia':'yellow',
    'Europe':'red',
    'Africa': 'black',
    'North America':'blue',
    'South America':'g',
    'Oceania':'c'
}
colors = world.Continent.map(map_dict)
size = world.Population/1e6
plt.xscale('log')
plt.ylabel('population', size= 20)
plt.xlabel('GDP per capita',size=20)
plt.text(1000,800000000,'no reation between GDP and population',size =20)
plt.xlim(world.GDP_per_capita.min(), world.GDP_per_capita.max())
plt.scatter( world.GDP_per_capita, world.Population,  c=colors, s=size, alpha=0.5)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
#円グラフ
# Control size  by default it takes the least element size (set it always)
fig = plt.figure(figsize =(8,8))
plt.title('Pie chart',size =15)
# 各ブブンのラベル　 labels
labels =['Python','Java','C++','Ruby']

# 各部分の数字　比率
values = [20,30,40,10]

#各部分の色
colors =['orange', 'red', 'blue', 'magenta']

# 　決まったラベルを中心から離れる　explode any  the labels
explode = [0,0.3,0,0]

# 影を付ける shadow  
#show the ratios as percentage  autopct 各部分の大きさを%で表示する
auto = '%0.1f%%'

# ｐｃｔdistance : autopctの数値が中心からどれぐらい離れる
dist = 0.4

#円の大きさ radius
r =1.5
# set labels text size textprops
tsize = {'fontsize':20}
plt.pie(values, labels=labels, colors=colors, shadow=True,
        explode=explode, autopct=auto, pctdistance=dist,
        radius=r, textprops=tsize)

plt.show()

```

```python
#空白円と回転
import matplotlib.pyplot as plt
import numpy as np
#円グラフ
# Control size  by default it takes the least element size (set it always)
fig = plt.figure(figsize =(6,6))
plt.title('Pie chart',size =15)
# 各部分　 labels
labels =['Python','Java','C++','Ruby']

# 各部分の数字　比率
values = [20,30,40,10]

#各部分の色
colors =['orange', 'red', 'blue', 'magenta']

# wedgeprops ={'linewidth':, 'width': ,'edgecolor': }
#linewidth 枠線の太さ
#width　円中心の空白
#edgecolor　枠線の色
exp = [0.1,0,0,0]
auto = '%0.1f%%'
rad = 2
wp = {'linewidth':4, 'width':0.6,'edgecolor':'green'}
#textprops は文字のスタイルを設定する
ftsize = {'fontsize':16,'color':'purple', 'weight':'bold'}
pdist = 0.8
rot = 90 
plt.pie(values,labels=labels, colors=colors, 
        shadow=True, explode=exp, autopct=auto,
        pctdistance=pdist, radius=rad, wedgeprops=wp,
        startangle=rot, textprops=ftsize)
plt.show()
```

```python
#必要ライブラリの読み込み
import pandas as pd

#データの読み込み
df = pd.read_csv("International_football_results.csv")
```

```python
#演習問題1

#1.列名が"tournament"の列を取り出せ。
df["tournament"]


#2.行番号100行目以上6000行目未満を取り出せ。
df[100:6000]

#3.列名が"date","city","coutry"となっている列を3列まとめて取り出せ。
df[["date","city","country"]]

#4."Brazil"がhome_teamとなっている試合を取り出せ。
df[df.home_team == "Brazil"]

#5."Brazil"が出場している試合を全て取り出せ。
df[(df.home_team == "Brazil") | (df.away_team == "Brazil")]

#6.2003年に行われた試合を全て取り出せ。
df[df.date.str.contains("2003")]
```

```python
#演習問題2

#1. 試合2チームの総合得点の列、point_addを追加せよ
point_add = df["home_score"] + df["away_score"]
df["point_add"] = point_add

#2.イングランド(England)が出場した試合を全て抜き出し新たな変数に代入せよ
Eng = df[(df[["home_team","away_team"]] == "England").any(axis = 1)]
print(Eng.head())
#3. 2で作成したデータフレームをcsvファイルへ書き出せ
Eng.to_csv("England.csv",index = False)
```

```python
#総合演習問題
df = pd.read_csv("sample-data.csv")
#1. データの基本情報を確認せよ。（行数、列数、初めの5行など）
print(df.info())
print(df.head())

#2.データには欠損値がある。欠損値のある行を除け
print(df.dropna())

#3.性別（Gender）の列をダミー変数化してデータに付け加えよ
df = pd.concat([df.drop(["Gender"],axis=1),pd.get_dummies(df["Gender"])],axis=1)
print(df.head())
#4.病気（Disease）が1の人のみを抜き出せ
print(df[df.Disease == 1])

#データを学習用と検証用に8対2の割合で分割せよ。
from sklearn.model_selection import train_test_split
train_df, test_df = train_test_split(df,test_size=0.2)
```

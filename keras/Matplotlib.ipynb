```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2*np.pi,0.1)
y = np.sin(x)
plt.title('sin graph')
plt.grid(True)
plt.plot(x,y, c= 'r')
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set(xlabel='x axis', ylabel='y axis',title='sine function with ax object' )
ax.grid()
plt.show()
```

```python
#import libraries
from tensorflow.python.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
from tensorflow.python.keras.preprocessing import image
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np
import sys

#Enter model VGG16
model = VGG16(weights = 'imagenet')
# function for image recognition
def predict(filename,featuresize):
    img = image.load_img(filename, target_size=(224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    preds = model.predict(preprocess_input(x))
    results = decode_predictions(preds, top=featuresize)[0]
    return results

# function for vizualisation
def showimg(filename, title, i):
    im = Image.open(filename)
    im_list = np.asarray(im)
    plt.subplot(2,5,i)
    plt.title(title)
    plt.axis('off')
    plt.imshow(im_list)
    

#Define the image
filename = "Train/train/cat.2591.jpg"
plt.figure(figsize=(20,10))
for i in range(1):
    showimg(filename, "query", i+1)
plt.show()
results = predict(filename, 10)
for result in results :
    print(result)
    
#Image recognition
filename  = "Train/train/dog.4035.jpg"
plt.figure(figsize = (20,10))
for i in range(1):
    showimg(filename,"query", i+1)
plt.show()
results = predict(filename, 10)
for result in results:
    print(result)

    
```

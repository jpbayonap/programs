```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def dif(t,c0,c1):
    return  c0+(t*c1) 
```

```python
guess= [0.01, 0.52]
x =np.array([1.489958479,1.549003262,1.908485019])
y= np.array([0.390935107,0.421603927,0.598790507])
```

```python
from scipy.optimize import curve_fit

c,cov = curve_fit(dif,x,y,guess)
c
```

```python
n = len(x)
y_fit= np.empty(n)

for i in range(n) :
    y_fit[i] = dif(x[i],c[0],c[1])

```

```python

plt.ylabel(r'$log(\Delta E/E )$')
plt.xlabel('log(E) Kev')
plt.title('Energy  peack resolution vs energy ')       

plt.scatter(x,y,label =' Enery resolution  ')
plt.plot(x,y_fit,'r',label = 'linear regression'+r'$ %.2f + %.2f x$'%(c[0],c[1]))
plt.legend()
plt.savefig('/home/juan/Desktop/rad_ln.ps', format= 'ps')
plt.show()

```

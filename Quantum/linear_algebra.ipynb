```python
import matplotlib.pyplot as plt
import numpy as np
from qiskit import *
from qiskit.visualization import plot_bloch_vector
```

```python
plt.figure()
ax = plt.gca()
X = []
Y = []
r =20
for i in range(12):
    X.append([r*np.sin((i+1)*np.pi/(6))])
    Y.append([r*np.cos((i+1)*np.pi/(6))])
for i in range(12):
     ax.quiver(X[i],Y[i],angles= "xy", scale_units="xy", scale = 1)

ax.set_xlim([-r,r])
ax.set_ylim([-r,r])
plt.draw()
plt.show()
```

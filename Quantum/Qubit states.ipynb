```python
from qiskit import QuantumCircuit, execute, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import pi, sqrt
```

```python
%config InlineBackend.figure_format = 'svg'
```

```python
qc = QuantumCircuit(1) #Create one qubit circuit
```

```python
#use initialize to change the state of the qubit
from qiskit.extensions import Initialize
initial_state = [1,0] # define the initial state as |0>
init_op = Initialize(initial_state) # Create initialization operation for  state |0>
qc.append(init_op,[0]) # Apply initialization 
qc.draw()
```

```python
backend = Aer.get_backend('statevector_simulator') # How to simulate the qubit
```

```python
result = execute(qc, backend).result() # Do simulation ,get result
```

```python
out_state = result.get_statevector() # get the  final statevector from  result
print(out_state)
```

```python
qc.measure_all()
qc.draw()
```

```python
# get counts for each result
result = execute(qc, backend).result()
counts = result.get_counts()
plot_histogram(counts)
```
